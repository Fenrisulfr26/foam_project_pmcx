10cm_data 可视化分析

```python
from pathlib import Path
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib qt

data_dir = Path(r'F:/OneDrive/foam_imaging_project/10cm_data')
image_dir = data_dir / 'images'

npy_files = sorted(data_dir.glob('*.npy'))
bmp_files = sorted(image_dir.glob('*.bmp'))

print(f'Data folder: {data_dir}')
print(f'Found {len(npy_files)} npy files and {len(bmp_files)} bmp images.')

dataset_summary = []
loaded_arrays = {}
for file_path in npy_files:
    arr = np.load(file_path)
    loaded_arrays[file_path.stem] = arr
    dataset_summary.append({
        'name': file_path.name,
        'shape': arr.shape,
        'dtype': str(arr.dtype),
        'min': float(np.min(arr)),
        'max': float(np.max(arr)),
        'mean': float(np.mean(arr)),
    })

for item in dataset_summary:
    print(f"{item['name']}: shape={item['shape']}, dtype={item['dtype']}, min={item['min']:.3f}, max={item['max']:.3f}, mean={item['mean']:.3f}")

fig, axes = plt.subplots(2, 3, figsize=(16, 9))
axes = axes.ravel()

# IRF
irf = loaded_arrays['IRF_gain0.7_timebin15ps_2000bins_080426']
axes[0].plot(irf[:])
axes[0].set_title('IRF (first 1000 bins)')
axes[0].set_xlabel('Time bin')
axes[0].set_ylabel('Counts')

# Sensitivity map
sens = loaded_arrays['sensitivity_map_gain0.7_85x85mm_51x51points_binWidth15ps_expo0.1sec_binNum2000_40deg__080426']
sens_2d = sens.sum(axis=2) if sens.ndim == 3 else sens
im1 = axes[1].imshow(sens_2d, cmap='hot')
axes[1].set_title('Sensitivity map')
plt.colorbar(im1, ax=axes[1], fraction=0.046)

# 31x31 dataset example
data_31 = loaded_arrays['H_gain0.7_5x5cm_31x31points_pol110deg_expo1.0sec_binWidth15ps_binNum2000_080426']
axes[2].plot(data_31[15, 15, 276:1109])
axes[2].set_title('H 31x31 center TPSF')
axes[2].set_xlabel('Time bin')

# 3x3 dataset overview
data_3 = loaded_arrays['H_gain0.7_5x5cm_3x3points_pol110deg_expo6.0sec_binWidth15ps_binNum2000_080426']
im2 = axes[3].imshow(data_3.sum(axis=2), cmap='viridis')
axes[3].set_title('H 3x3 summed intensity')
plt.colorbar(im2, ax=axes[3], fraction=0.046)

# noabs reference trace comparison
ref_31 = loaded_arrays['noabs_gain0.7_5x5cm_31x31points_pol110deg_expo1.0sec_binWidth15ps_binNum2000_080426']
ref_3 = loaded_arrays['noabs_gain0.7_5x5cm_3x3points_pol110deg_expo6.0sec_binWidth15ps_binNum2000_080426']
axes[4].plot(ref_31[15, 15, 276:1109], label='31x31 center')
axes[4].plot(ref_3[1, 1, 276:1109], label='3x3 center')
axes[4].set_title('No-abs reference comparison')
axes[4].legend()

# image preview
img = cv2.imread(str(image_dir / 'H_exp2685ms.bmp'), cv2.IMREAD_GRAYSCALE)
axes[5].imshow(img, cmap='gray')
axes[5].set_title('H target image')
axes[5].axis('off')

plt.tight_layout()
plt.show()
```
